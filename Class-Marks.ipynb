```python
import pandas as pd
df = pd.read_csv("class_marks.csv")
df
```

# Importing pandas and reading data from CSV file

```python
#Labelling qus by adding sub-qus into single column name
df['Qus1'] = df['Q1aM4'] + df['Q1bM6']
df['Qus2'] = df['Q2aM6'] + df['Q2bM4']
df['Qus3'] = df['Q3aM5'] + df['Q3bM5']
df['Qus4'] = df['Q4aM3'] + df['Q4bM7']
df['Qus5'] = df['Q5M10']
df['Qus6'] = df['Q6aM4'] + df['Q6bM6']

df.drop(['Q1aM4', 'Q1bM6', 'Q2aM6', 'Q2bM4', 'Q3aM5', 'Q3bM5', 'Q4aM3', 'Q4bM7', 'Q5M10', 'Q6aM4', 'Q6bM6'], axis=1, inplace=True)
df.head()
```

# Creating new columns for questions and dropping old ones for the easy analysis of data

```python
#Finding top 5 members for each question
top_5_per_question = {}
for column in df.columns[1:]:
    top_5_per_question[column] = df.nlargest(5, column)
print(top_5_per_question["Qus1"])
print(top_5_per_question["Qus2"])
print(top_5_per_question["Qus3"])
print(top_5_per_question["Qus4"])
print(top_5_per_question["Qus5"])
print(top_5_per_question["Qus6"])
```

# Finding top 5 scores per question and displaying results which is useful for identifyong how the students were addressing questions as of from each topic

```python
#Missing values for each Question
df.isnull().sum()
```

# Counting the number of null values in each column which is esential for analysing which question is not attempted by many students, that helps in analyzing which topic furthur needed to be discussed.

```python
#Filling the missing values
df.fillna(df.mean(), inplace=True)
df
```

# Filling missing values in the dataframe with column means with null values 


```python
#Rounding-off the Scores
columns_to_round = ['Qus1', 'Qus2', 'Qus3', 'Qus4', 'Qus5', 'Qus6']
df[columns_to_round] = df[columns_to_round].round()
df.head()
```

# Rounding values in specific question columns

```python
#Averge for each Questions
average_scores = df[['Qus1', 'Qus2', 'Qus3', 'Qus4', 'Qus5', 'Qus6']].mean()
print(average_scores)
```

# Calculating and printing average scores for each question

```python
#toppers for each Questin
toppers = {}
for col in ['Qus1',
            'Qus2',
            'Qus3',
            'Qus4',
            'Qus5',
            'Qus6']:
    toppers[col] = df[df[col] == df[col].max()].index.tolist()

print("Toppers as per each Question:", toppers)
```

# Identifying and printing toppers for each question based on maximum scores

```python
#Bar-Chart that how scores Awarded  
import matplotlib.pyplot as plt
df['Total'].plot(kind='hist', bins=10, color='maroon', edgecolor='blue', alpha=0.7)
plt.title("How scores are awarded")
plt.xlabel("Scores")
plt.ylabel("Reach")
plt.show()
```
